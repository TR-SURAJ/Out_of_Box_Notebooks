- Lets try to find association between categorical features and the target variable

```python
from sklearn.feature_selection import SelectKBest
```

```python
cat_feat = ['data_channel_is_lifestyle','data_channel_is_bus','data_channel_is_socmed','data_channel_is_tech',
           'data_channel_is_world','weekday_is_monday','weekday_is_tuesday','weekday_is_wednesday','weekday_is_thursday','weekday_is_friday','weekday_is_saturday',
           'weekday_is_sunday','is_weekend']
X = df[cat_feat]
y = df['shares']
```

#### Chi2 test

```python
from sklearn.feature_selection import chi2
fs = SelectKBest(score_func=chi2, k='all')
fs.fit(X, y)
Xt = fs.transform(X)
for i,j in zip(range(len(fs.scores_)),X.columns):
    print('Feature %s: %f' % (j, fs.scores_[i]))

plt.barh(X.columns, fs.scores_)
plt.show()
```

- We can observe that among the 13 categorical features data_channel_is_world(data channel World),is_weekend(Was the article published on the weekend?),weekday_is_saturday(Was the article published on a Saturday?) are top 3 predicters contributing in predicting target shares as they have strong association. These 3 are followed by data_channel_is_socmed,data_channel_is_lifestyle
- This actually make sense because the number of shares is pretty more during saturday/sunday as it would be holiday for most of the working professionals and have time to read and share the news to the fellow friends. 
- The world news is quite more popular than others as it more generalized news covering all the aspects across different parts of the world

#### ANOVA - f_classify

```python
from sklearn.feature_selection import f_classif
fs = SelectKBest(score_func=f_classif, k='all')
fs.fit(X, y)
Xt = fs.transform(X)
for i,j in zip(range(len(fs.scores_)),X.columns):
    print('Feature %s: %f' % (j, fs.scores_[i]))

plt.barh(X.columns, fs.scores_)
plt.show()
```

- ANOVA test gave similar results as well just with different scoring startegy (F-Scores)

##### There were few of the continous features for which the correlation with the target was zero or nearly zero. Lets remove them as they wouldn't contribute in predicting y. I have considered features with correlation <= 0.001

##### Excluding cat_feat we can observe the distributions of the continuous variables which is either positive skewed or negative skewed. Lets try to make skewed features to gaussian which make our model work better.

```python
len(truncdf.columns)
```

```python
cat_feat = ['data_channel_is_lifestyle','data_channel_is_bus','data_channel_is_socmed','data_channel_is_tech',
           'data_channel_is_world','weekday_is_monday','weekday_is_tuesday','weekday_is_wednesday','weekday_is_thursday','weekday_is_friday','weekday_is_saturday',
           'weekday_is_sunday','is_weekend']
```

```python
truncdf.drop(cat_feat,axis = 1)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(truncdf,y,test_size = 0.2)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from sklearn.preprocessing import QuantileTransformer
import warnings
warnings.filterwarnings('ignore')
```

```python
for i in truncdf.columns:
    trans = QuantileTransformer(n_quantiles=100, output_distribution='normal')
    X_train.loc[:,i] = trans.fit_transform(X_train[[i]])
    X_test.loc[:,i] = trans.transform(X_test[[i]])
```

#### Ignore warnings 

```python
X_train.head()
```

```python
X_train.hist(figsize=(16,16))
plt.show()
```

```python
X_test.hist(figsize=(16,16))
plt.show()
```

Observe the change in distribution for continuous variables

```python
y.hist()
```

Even target variable has to be transformed because of high negative skew

```python
y_train_new = np.log(y_train)
y_test_new = np.log(y_test)
```

```python
y_train_new.hist()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error

lr = LinearRegression()
lr.fit(X_train,y_train_new)

ypredtr = lr.predict(X_train)
ypred = lr.predict(X_test)

print('train r2',r2_score(y_train_new,ypredtr))
print('train rmse',np.sqrt(mean_squared_error(y_train_new,ypredtr)))

print('test r2',r2_score(y_test_new,ypred))
print('test rmse',np.sqrt(mean_squared_error(y_test_new,ypred)))
```

#### RMSE for both train and test is quite the same, no overfitting the data

#### Power transformers

```python
X = df.iloc[:,1:-1]
y = df.iloc[:,-1]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import PowerTransformer

std = StandardScaler()
X_train = std.fit_transform(X_train)
X_test = std.transform(X_test)

power = PowerTransformer(method='yeo-johnson')
X_train = power.fit_transform(X_train)
X_test = power.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error

lr = LinearRegression()
lr.fit(X_train,y_train_new)

ypredtr = lr.predict(X_train)
ypred = lr.predict(X_test)

print('train r2',r2_score(y_train_new,ypredtr))
print('train rmse',np.sqrt(mean_squared_error(y_train_new,ypredtr)))

print('test r2',r2_score(y_test_new,ypred))
print('test rmse',np.sqrt(mean_squared_error(y_test_new,ypred)))
```
