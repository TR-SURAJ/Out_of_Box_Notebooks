```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
import numpy as np
```

```python
# read the excel file
data = pd.read_excel("Interview_questions.xlsx", engine='openpyxl')
```

```python
data.head(20)
```

```python
data.dtypes
```

```python
target_date = "2020-10-22"
```

```python
days = data.loc[0,'date'] - pd.to_datetime(target_date)
```

```python
days.days
```

```python
data['diff'] = data['date'].apply(lambda x: (x - pd.to_datetime(target_date)).days)
```

```python
data.head()
```

```python
data['fillrate'] = data['rcvval'] / data['orderval']
```

```python
data.head()
```

```python
filter_method = lambda x: '-1 to -10' if(x >= -10 and x <= -1) else '0 to 2' if(x >= 0 and x <= 2) else '3 to 7' if(x >= 3 and x <= 7) else None
data['diff_bin'] = data['diff'].apply(filter_method)
```

```python
data.head()
```

```python
data = data.drop(['diff','fillrate'],axis =1)
```

```python
data.pivot_table(index=['format','division_name'], columns='diff_bin').swaplevel(axis=1).sort_index(1)
```

```python
pivot_data = data.pivot_table(index=['format','division_name'], columns='diff_bin').swaplevel(axis=1).sort_index(1)
```

```python
pivot_data
```

```python
df1 = pivot_data.groupby(['format']).sum()
df1.index = pd.MultiIndex.from_arrays([df1.index + '_total', len(df1.index) * ['']])
print (df1)
```

```python
pd.concat([pivot_data, df1]).sort_index()
```

```python
finaldf = pd.concat([pivot_data, df1]).sort_index()
```
